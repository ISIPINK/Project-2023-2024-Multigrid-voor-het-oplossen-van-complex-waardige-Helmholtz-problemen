# test relax

```python
def test_relax_solving(relax,n,iterations):
    A = diags([-1, 4, -1], [-1, 0, 1], shape=(n, n))
    f = np.array(range(n))
    u =  np.zeros(n)
    for _ in range(iterations):
        u = relax(A, f, u)
    usol = spsolve(A,f) 
    plt_vecs([u-usol,u],["error","sol"])

def test_relax_smoothing(relax,n,iterations):
    A = diags([-1, 4, -1], [-1, 0, 1], shape=(n, n))
    f = np.array(range(n))
    usol = spsolve(A,f) 
    u =  20*wave_basis_1D(n+1,n//2)
    errors = [u-usol]
    for _ in range(iterations):
        u = relax(A, f, u)
        errors.append(u-usol)
    plt_vecs(errors,["error"]*len(errors))

omega = 2/3
test_relax_solving(partial(wjacobi,omega=omega),n=20,iterations=60)
test_relax_smoothing(partial(wjacobi,omega=omega),n=20,iterations=60)
```

# test interpolate and restrict

```python
import numpy as np
def test_recursive_restriction(depth,v,restrict):
    vecs= [v]
    for _ in range(depth):
        try:
            v = restrict(v)
        except:
            break
        vecs.append(v)
    return vecs
```

```python
vecs = test_recursive_restriction(7,wave_basis_1D(2**8,5),simple_restrict)
plt_vecs(vecs)
```

```python
vecs = test_recursive_restriction(7,wave_basis_2D(2**8,3,3),simple_restrict2D)
plt_vecs2D(vecs)
```

```python
def test_interpolate_restrict_error(n,vecs,interpolate,restrict):
    vecs_reconstructed = [interpolate(restrict(v))-v for v in vecs ]
    norms = [norm(v)/norm(w) for v,w in zip(vecs_reconstructed,vecs) if norm(w)>0 ]  
    plt.plot(range(len(norms)),norms)
    plt.title("reconstruction error %norm of vecs")
    plt.grid(True)
    plt.show()
    return vecs_reconstructed
```

```python
n = 50
wbasis_1D = [wave_basis_1D(n,k) for k in range(n)]
test_interpolate_restrict_error(n,wbasis_1D,simple_interpolate,simple_restrict)
```

```python
n = 20
wbasis_2D = [wave_basis_2D(n,k,k) for k in range(n)]
recs = test_interpolate_restrict_error(n,wbasis_2D,simple_interpolate2D,simple_restrict2D)
plt_vecs2D(recs)
```

# test Vcycles

```python
from functools import lru_cache
def laplace1D_test(n=32,iterations=1,recursion_depth=3,cache=False):
    A = helmholtz1D(n,0)
    u_exact = wave_basis_1D(n,1)
    f = A @ u_exact 
    u_approx = np.zeros(n-1)
    cache_depth = recursion_depth if cache else 0

    errors = [u_exact-u_approx] 
    # Solve using the V-cycle method
    for _ in range(iterations):
        u_approx = geoVcycle(
            mat = lru_cache(cache_depth)(partial(helmholtz1D,sigma=0)), 
            f= f, 
            u = u_approx, 
            nu1=1,
            nu2=1,
            relax=partial(wjacobi,omega= 2/3), 
            restrict=simple_restrict, 
            interpolate=simple_interpolate, 
            recursion_depth=recursion_depth)
        errors.append(u_exact-u_approx)
    plt_vecs(errors,[f"errors after {j} vcycles" for j in range(len(errors))]) 
    
laplace1D_test(2**7,20,6,cache=True)
```

```python
def laplace2D_test(n=32,iterations=1,recursion_depth=3,cache=False):
    A = helmholtz2D(n,0)
    u_exact = wave_basis_2D(n,1,1)
    f = A @ u_exact 
    u_approx = np.zeros((n-1)**2)
    cache_depth = recursion_depth if cache else 0

    errors = [u_exact-u_approx] 
    # Solve using the V-cycle method
    for _ in range(iterations):
        u_approx = geoVcycle(
            mat = lru_cache(cache_depth)(partial(helmholtz2D,sigma=0)), 
            f= f, 
            u = u_approx, 
            nu1=1,
            nu2=1,
            relax=partial(wjacobi,omega= 2/3), 
            restrict=simple_restrict2D, 
            interpolate=simple_interpolate2D, 
            recursion_depth=recursion_depth,
            dimensions =2)
        errors.append(u_exact-u_approx)
    plt_vecs2D(errors,[f"errors after {j} vcycles" for j in range(len(errors))]) 
    
laplace2D_test(2**(2*4),10,4,cache=False)
```
